```python
# import packages
import uproot
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the ROOT file with your data
file = uproot.open("DarkMatter_MonoPhoton.root")

print(file.keys())
```

```python
## Fill in the correct tree name below
events = file["Events"]

## inspect the available branches 
events.keys()
```

```python
# Read the branches into an array
## The variable names are plecahorly defined in the root file. You can use the `events.keys()` command to see all available branches.
photon_pt = events["Photon_pt"].array()
photon_eta = events["Photon_eta"].array()
photon_phi = events["Photon_phi"].array()

electron_pt = events["Electron_pt"].array()
muon_pt = events["Muon_pt"].array()

met_pt = events["MET_pt"].array()
met_phi = events["MET_phi"].array()
```

```python
## This is a mono photon analysis, so we will require that there are no electrons or muons in the event.
hasPhoton = ak.num(photon_pt) > 0

photon_pt = photon_pt[hasPhoton]
photon_eta = photon_eta[hasPhoton]
photon_phi = photon_phi[hasPhoton]

electron_pt = electron_pt[hasPhoton]
muon_pt = muon_pt[hasPhoton]

met_pt = met_pt[hasPhoton]
met_phi = met_phi[hasPhoton]
```

```python
## lets take the leading photon in each event
leadPhotonPt = photon_pt[:,0]
leadPhotonPhi = photon_phi[:,0]
```

```python
## selecting a photon 
# you can chnage the threshold to see how it affects the number of events
mask = leadPhotonPt > 150

leadPhotonPt = leadPhotonPt[mask]
leadPhotonPhi = leadPhotonPhi[mask]

electron_pt = electron_pt[mask]
muon_pt = muon_pt[mask]

met_pt = met_pt[mask]
met_phi = met_phi[mask]
```

```python
## Since we are looking for dark matter, we will require that the missing transverse energy is greater than 150 GeV.
mask = met_pt > 200

leadPhotonPt = leadPhotonPt[mask]
leadPhotonPhi = leadPhotonPhi[mask]

electron_pt = electron_pt[mask]
muon_pt = muon_pt[mask]

met_pt = met_pt[mask]
met_phi = met_phi[mask]
```

```python
# veto electrons and muons 
electronVeto = ak.num(electron_pt) == 0
# write the code to veto muons here
```

```python
# you can also combine both vetoes 
mask = electronVeto & muonVeto

leadPhotonPt = leadPhotonPt[mask]
leadPhotonPhi = leadPhotonPhi[mask]

met_pt = met_pt[mask]
met_phi = met_phi[mask]
```

```python
# Plot the transverse momentum of the leading photon
plt.hist(
    ak.to_numpy(leadPhotonPt),
    bins=40
)

plt.xlabel("Leading photon $p_T$ [GeV]")
plt.ylabel("Events")
plt.show()
```

```python
# plot the missing transverse energy
plt.hist(
    ak.to_numpy(met_pt),
    bins=40
)

plt.xlabel("Missing transverse energy [GeV]")
plt.ylabel("Events")
plt.show()
```

```python
## Calculate delta phi between the leading photon and the missing transverse energy
deltaPhi = np.abs(leadPhotonPhi - met_phi)

deltaPhi = ak.where(
    deltaPhi > np.pi,
    2*np.pi - deltaPhi,
    deltaPhi
)
```

```python
# plot the delta phi distribution
plt.hist(
    ak.to_numpy(deltaPhi),
    bins=30
)

plt.xlabel(r"$\Delta\phi(\gamma,\mathrm{MET})$")
plt.ylabel("Events")
plt.show()
```
